Let's experiment with column "10. What do you expect to get out of this course?" to see if we can determine most popular leisure time actvities.

````python
list_answers = df["9. What do you enjoy doing in your leisure time?"].tolist()

list_answers
````

We will start with an empty dictionary like this:
```python
word_freq = {}
```

We hope to end up with something like this:
```python
word_freq{
    "movies" : 3,
    "reading": 10,
    "sports": 3
}
```

````python
word_freq = {}

for answer in list_answers:
    list_words = answer.split()
    for word in list_words:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1

word_freq
````

## Assignment 03 Part 2

Process column "11. What is your long-term career goal?" to find out the most popular career goals of the class. User Python code to parse the responses and count occurences of words, then use your naked eyes to make determination. Write your observation/analysis using Markdown.

````python
# Extracting the responses from the specified column
career_goals_list = df["11. What is your long-term career goal?"].tolist()

# Display the list of career goals
career_goals_list
````

````python
# Extracting the responses from the specified column
career_goals = df["11. What is your long-term career goal?"].tolist()

# Initializing a dictionary to count word occurrences
word_freq_career = {}

# Processing the responses
for goal in career_goals:
    list_words = goal.split()
    for word in list_words:
        if word in word_freq_career:
            word_freq_career[word] += 1
        else:
            word_freq_career[word] = 1

# Display the word frequency dictionary
word_freq_career
````

##**Observations**:

 Important results showed a strong desire for professions that are driven by data (such as "data" and "analyst"). A wide range of goals were expressed in the responses, including management and education, with many students thinking about pursuing further education.
