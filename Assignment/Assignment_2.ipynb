## Step 2 - Explore Student Age

```python
yob_list = list(df["8. My Birth Year"])
yob_list
```

```python
cleansed_yob_list = []  # Empty list
cleansed_yob_list       # the same as print(cleansed_yob)
```

```python
for yob in yob_list:
    cleansed_yob = yob[-4:]
    cleansed_yob_list.append(cleansed_yob)

cleansed_yob_list
```

```python
age_list = []

for yob in cleansed_yob_list:
    age = 2024 - int(yob)
    age_list.append(age)

age_list
```

```python
min(age_list)
```

```python
max(age_list)
```

```python
total_age = 0

for age in age_list:
    total_age = total_age + age

total_age
```

```python
average = total_age / len(age_list)

round(average)
```

```python
total_age = 0

for age in age_list:
    if age >= 35 or age <= 22:
        print("Age", age, "is skipped")
    else:
       total_age = total_age + age

print("Total age is", total_age)

average = round(total_age / len(age_list))

print("Average is", average)
```

```python
total_age = 0

for age in age_list:
    if age >= 35 or age <= 22:
        print("Age", age, "is skipped")
#        continue

    total_age = total_age + age   # the age out of range was not skipped without the "continue"

print("Total age is", total_age)

average = round(total_age / len(age_list))

print("Average is", average)
```

## Step 3 - Explore Educational Background

Compute the distribution of student edu background


```python
edu_list = list(df["1. My primary educational background"])

edu_list
```

```python
df_edu_dist = df["1. My primary educational background"].value_counts()
df_edu_dist
```

```python
df_edu_dist = df_edu_dist.reset_index()
df_edu_dist
```

```python
fig = px.bar(
    df_edu_dist,
    x="1. My primary educational background",
    y="count",
    color="1. My primary educational background",
    title="UMBC DATA 690 Fall 2024 Student Education Distribution"
)

fig.show()
```

### Home Work

```python
# Create the pie chart
fig = px.pie(
    df_edu_dist,
    names='1. My primary educational background',  # Column for educational backgrounds
    values='count',  # Column for counts
    title="UMBC DATA 690 Fall 2024 Student Education Distribution"
)

```

## Step 5 - Bonus

distribution of "12. Other courses taking this semester" with Bar and Pie Chart

```python
df["12. Other courses taking this semester"]
```

function to operate on srings:

To take care of cases. - convert to upper() or to lower()

```python
# Distribution of courses
df_courses_dist = df["12. Other courses taking this semester"].value_counts().reset_index()
df_courses_dist.columns = ["Course", "Count"]

# Bar chart
fig_bar = px.bar(
    df_courses_dist,
    x="Course",
    y="Count",
    color="Course",
    title="Distribution of Other Courses Taken This Semester - Bar Chart"
)
fig_bar.show()
```

```python
# Pie chart
fig_pie = px.pie(
    df_courses_dist,
    names="Course",
    values="Count",
    title="Distribution of Other Courses Taken This Semester - Pie Chart"
)
fig_pie.show()
```
